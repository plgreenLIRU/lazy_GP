```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
from lazy_GP import GP
```

Generate 1D regression problem

```python
np.random.seed(42)  # for reproducibility
sigma = 0.05
N = 50
X = np.linspace(0, 10, N)[:, np.newaxis]
X_star = np.linspace(0, 10, 100)[:, np.newaxis]
y = np.sin(X[:, 0]) + sigma * np.random.randn(N)
y_star = np.sin(X_star[:, 0]) + sigma * np.random.randn(len(X_star))
```

```python
fig, ax = plt.subplots()
ax.plot(X, y, 'o')
ax.grid()
plt.show()
```

Fit GP

```python
m = GP()
theta = np.array([3.])
m.train(X=X, y=y, theta=theta, sigma=sigma)
```

Make predictions

```python
y_star_pred = m.predict(X_star)
```

Results

```python
fig, ax = plt.subplots()
ax.plot(X, y, 'o', label='Data', color='black')
ax.plot(X_star, y_star_pred, '-', color='red', label='Prediction')
ax.legend()
ax.grid()
plt.show()
```
